```python
def distance_pixel2pixel (a, b):
  m=0
  for i in range (0,3):
    m+=(a[i]-b[i])**2
  distance= m**0.5
  return distance
```

```python
### 수성0,금성1,지구2,목성3,천왕성4
data = []

for i in range(5): 
  for j in range(2):
    file_path = f'/gdrive/My Drive/prj/{i}/{j}.png'
    img = cv.imread(file_path)
    img = cv.resize(img,(100,100))
    data.append([img, i])
    cv2_imshow(img)

print(len(data))
```

```python
file_path = f'/gdrive/My Drive/prj/4/2.png'
input_img = cv.imread(file_path)
input_img = cv.resize(input_img, (100,100))
cv2_imshow(input_img)
```

```python
label_count = [0,0,0,0,0]

for ii in range(3):
  for j in range(3):
    x, y = 30+20*ii, 30+20*j
    distance = []
    distance_data = []
    for i in range (0,10):
      planet = data[i][0][x][y]
      d = distance_pixel2pixel(planet, input_img[x][y])
      distance_data.append([distance_pixel2pixel(planet, input_img[x][y]), data[i][1]])
      distance.append(distance_data[i][0])
    sorted_index = np.array(distance).argsort()
    cloest_index = sorted_index[0]
    label_count[distance_data[cloest_index][1]]+=1
# print([k/9 for k in label_count])
print(label_count)
```
